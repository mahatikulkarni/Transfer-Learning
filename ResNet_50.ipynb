**ResNet** - *ResNet, short for Residual Networks is a classic neural network used as a backbone for many computer vision tasks. This model was the winner of ImageNet challenge in 2015. The fundamental breakthrough with ResNet was it allowed us to train extremely deep neural networks with 150+layers successfully.*

image.png

```python
#Load the modules

from keras.applications.imagenet_utils import preprocess_input, decode_predictions
from keras.models import load_model
from keras.preprocessing import image
from tensorflow.keras.applications import ResNet50
```

```python
#Define the model
model = ResNet50(weights='imagenet')
```

Dog image

```python
#Load an image in PIL format 

original = image.load_img("/content/drive/MyDrive/Colab Notebooks/Transfer Learning/1-dog.jpg", target_size=(224, 224))
```

```python
display(original)
```

```python
#Convert the image from PIL format to NumPy format

numpy_image = image.img_to_array(original)
```

```python
import numpy as np
#Convert the image from NumPy format to Batch format

image_batch = np.expand_dims(numpy_image, axis=0)
```

```python
#Pre-process the image by subtracting the mean value from each channel of the images in the batch.

processed_image = preprocess_input(image_batch, mode='caffe')
```

```python
#Get the classification result by calling the model.predict function

preds = model.predict(processed_image)
```

```python
#Convert the result to human readable labels, using the method decode_predictions method.

pred_class = decode_predictions(preds, top=1)
```

```python
print("predicted object is:")
print(pred_class[0][0][1])
print("\naccuracy")
print(pred_class[0][0][2]*100)
```
